```python
ratings_with_books.groupby('title')['rating'].count().sort_values(ascending=False).reset_index()
```

```python
rating_count = ratings_with_books.groupby('title')['rating'].count().sort_values(ascending=False).reset_index()
```

```python
rating_count.head()
```

```python
rating_count.rename(columns={'rating':'num_ratings'}, inplace=True)
```

```python
rating_count.head()
```

```python
rating_count.shape
```

```python
ratings = ratings_with_books.merge(rating_count, on='title')
```

```python
ratings.head()
```

```python
x = ratings['num_ratings'] >= 50
```

```python
x[x]
```

```python
final_ratings = ratings[ratings['num_ratings'] >= 50]
```

```python
final_ratings.head()
```

```python
final_ratings.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_ratings.shape
```

```python
final_ratings.isnull().sum()
```

```python
book_pivot = final_ratings.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.head()
```

```python
book_pivot.shape
```

```python
book_pivot.isnull().sum()
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```
